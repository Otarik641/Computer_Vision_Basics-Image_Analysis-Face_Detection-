```python
#import libraries

import cv2

#callback function for the mouse, Rectangle

def draw_rectangle(event,
                  x,
                  y,
                  flags,
                  param):

    #write your global param
    #check the Project computer vision - Image Basic
    #pt1, pt2, top_left_clicked, bottom_right_clicked
    
    global pt1, pt2, top_left_clicked, bottom_right_clicked
    
    #create an event for left button down and assignd it to event
    
    if event == cv2.EVENT_LBUTTONDOWN:
    
    #reset your rectangle
        
        if top_left_clicked == True and bottom_right_clicked == True:
            
        # Give 0 value to pt1 & pt2
        
            pt1 = (0,0)
            pt2 = (0,0)
        
        
        #give false value to your tracker
        
        top_left_clicked = False
        bottom_right_clicked = False
        
    #check the top_leftclicked if its false
    
    if top_left_clicked == False:
        pt1 = (x,y)
        top_left_clicked = True
        
        
    #check the bottom_right_clicked if it is false
    elif bottom_right_clicked == False:
        pt2 = (x,y)
        top_left_clicked = True
        

#pt1, pt2 Global variable
pt1 = (0,0)
pt2 = (0,0)

#Tracker are false

top_left_clicked = False
bottom_right_clicked = False

#Take a video
cap = cv2.VideoCapture(0)

#create a named window for the connections
cv2.namedWindow('Test')

#set a mouse callback
cv2.setMouseCallback('Test', draw_rectangle)


#time for magic
while True:
    ret, frame = cap.read()
    
    #based on the global variable draw the frame
    if top_left_clicked == True:
        cv2.circle(frame,
                  center=pt1,
                  radius=5,
                  color= (255,0,255),
                  thickness =-1)
        
        
    #Draw a rectangle on the frame
    if top_left_clicked == True and bottom_right_clicked == True:
        cv2.rectangle(frame,
                     pt1,
                     pt2,
                     (0,0,255),
                     3)
                 
    
    #show it
    cv2.imshow('frame', frame)
    
    
    # 27 ==esc button. You can use any other button like ord('q')
    #if you try to close the window pressing the x you ll get in trouble
    if cv2.waitKey(5) & 0xFF == 27:
        break
    
    
#Never forget first to release
#and then to destroy
cap.release()
cv2.destroyAllWindows()











    
    
        
        
        
        
        
```
