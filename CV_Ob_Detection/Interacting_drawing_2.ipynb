```python
#import libraries

import cv2

#callback function for the mouse, Rectangle

def draw_circle(event,
                  x,
                  y,
                  flags,
                  param):
    
    global center, clicked

    #get the Mouse click Down and up
    #and track the centre
    if event == cv2.EVENT_LBUTTONDOWN:
        center = (x,y)
        clicked = False
        
    if event == cv2.EVENT_LBUTTONUP:
        clicked = True
        
        
        
#zero drawing of the circle
center = (0,0)
clicked = False


#Take a video
cap = cv2.VideoCapture(0)

#create a named window for the connections
cv2.namedWindow('Testing')

#set a mouse callback
cv2.setMouseCallback('Testing', draw_circle)


#time for magic
while True:
    ret, frame = cap.read()
    
    #check if clicked is True
    
    if clicked:
        
        #draw a Circle on the frame
        cv2.circle(frame,
                  center=center,
                  radius=50,
                  color=(255,0,255),
                  thickness=3)
                 
    
    #show it
    cv2.imshow('Testing', frame)
    
    
    # 27 ==esc button. You can use any other button like ord('q')
    #if you try to close the window pressing the x you ll get in trouble
    if cv2.waitKey(3) & 0xFF == 27:
        break
    
    
#Never forget first to release
#and then to destroy
cap.release()
cv2.destroyAllWindows()











    
    
        
        
        
        
        
```
