```python
#import libraries

import cv2
import numpy as np

#variables
#True When the mouse button is Down
#False When the mouse button is up

drawing = False
ex = -1
ey = -1

#function
#x,y,flags,Params are feed from OpenCV automatically

def draw_rectangle(event,x,y,flags,params):
    
    global ex, ey, drawing
    
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ex,ey = x,y
        
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            cv2.rectangle(img,
                         (ex, ey),
                         (x,y,),
                         (255,0,255),
                         -1)
        
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        cv2.rectangle(img,
                     (ex, ey),
                     (x,y,),
                     (255,0,255),
                     -1)
#connect the function with the callback
img = np.zeros((512,512,3),np.int8)

cv2.namedWindow(winname='My_Window')

#callback
cv2.setMouseCallback('My_Window', draw_rectangle)

#using OpenCV to show the Image

while True:
    
    cv2.imshow('My_Window', img)
    
    if cv2.waitKey(5) & 0xFF == ord('q'):
        break
        
cv2.destroyAllWindows()






```
